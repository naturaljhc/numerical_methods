```python
with open("data/first_order_ODE_output.json", 'r') as file:
    data = json.load(file)
```

```python
t0 = data["t0"]
tend = data["tend"]
n = data["n"]

t = np.linspace(t0, tend, n+1)
y = np.array(data["data"])
```

```python
len(y)
```

```python
plt.plot(t, y, color = 'b')
# plt.plot(t, np.exp(t), color = 'r', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Approximation of Antiderivative using Riemann Sums')
plt.grid(True)
plt.show()
```
