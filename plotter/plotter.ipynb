```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open("../data/integration_output.json", 'r') as file:
    data = json.load(file)
```

```python
x0 = data["x0"]
xend = data["xend"]
n = data["n"]

x = np.linspace(x0, xend, n)
y = np.array(data["data"])
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Approximation of Antiderivative using Riemann Sums')
plt.grid(True)
plt.show()
```
